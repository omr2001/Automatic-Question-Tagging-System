```python
################################## Person 2 #################################################
# Reading Questions Dataset
Question_DF = pd.read_csv('/content/Questions.csv' , encoding='latin-1')
Question_DF.head()
```

```python
print(Question_DF.shape)
```

```python
# Dropping not needed columns
Updated_Question = Question_DF.drop(['OwnerUserId' ,  'CreationDate' , 'ClosedDate' , 'Score'] , axis =1)
# Separating by Spaces 
def comb_tags_func(x):
    list_tags = x.values
    tag_comb = ' '.join(list_tags)
    return tag_comb
# Grouping by ID of Tags
mapping_dict = Tags_DF.groupby('Id')['Tag'].apply(comb_tags_func)
```

```python
# Performing Same of Tgs
Updated_Question['tag'] = Updated_Question['Id'].map(mapping_dict)
Updated_Question.head()
```

```python
# Dropping Null Values 
Updated_Question.dropna(inplace=True)
# Splitting Data
Updated_Question['tag'] = Updated_Question['tag'].apply(lambda x : x.split())
Updated_Question.head()
```

```python
# Returning Content of Second Row 
Updated_Question['Body'].iloc[1]
```

```python
# Replacing HTML Tags of body
Updated_Question['Body']=Updated_Question['Body'].str.replace('<p>',' ')
import re
Updated_Question['Body']=Updated_Question['Body'].apply(lambda x : re.sub('(https?:\/\/)?([\da-z\.-]+)\.([a-z\.]{2,6})([\/\w \.-]*)' , ' ' , x))
Updated_Question['Body']=Updated_Question['Body'].str.replace('</p>',' ')
Updated_Question['Body']=Updated_Question['Body'].str.replace('\n',' ')
Updated_Question['Body']=Updated_Question['Body'].str.replace('</a>',' ')
```

```python
# Removing Stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words=set(nltk.corpus.stopwords.words('english'))
Updated_Question['Body']=Updated_Question['Body'].apply(lambda x : x.lower())
Updated_Question['Body'] = Updated_Question['Body'].apply(lambda x : ' '.join([w for w in x.split() if not w in stop_words]))
```

```python
# Replacing a href tag with space
Updated_Question['Body']=Updated_Question['Body'].str.replace('<a href=" ">','')
Updated_Question['Body'] = Updated_Question['Body'].apply(lambda x : ' '.join([w for w in x.split() if len(w)>3]))
Updated_Question['Body'].iloc[1]
```

```python
# Performing Stemming on Snowball Stemmer
import nltk
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
nltk.download('stopwords')

stemmer = SnowballStemmer('english')
stop_words=set(nltk.corpus.stopwords.words('english'))

Updated_Question['Title'] = [i.replace("?", "") for i in Updated_Question['Title']]
Updated_Question['Title'] = Updated_Question['Title'].apply(lambda x : ' '.join([w for w in x.split() if not w in stop_words]))
Updated_Question['Title'] = [stemmer.stem(i) for i in Updated_Question['Title']]
######################################### End Person 2 #########################################
```
